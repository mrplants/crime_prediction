```python
#join the columns 

####################################### do balance -> Food that humans have to eat 
# balance.Element.unique().tolist() # want just  'Food supply quantity (kg/capita/yr)' for now
# balance.Unit.unique().tolist()
balance2 = balance.loc[balance['Unit'] == '1000 tonnes'].copy()
balance2 = balance2.loc[balance2['Element'] == 'Food'].copy()  # http://www.fao.org/faostat/en/#data/FBS/report
balance2.rename(columns =  {'Element':'Food'}, inplace=True) #don't really need this, but good reminder that element is now food (but ws)
balance2 = balance2[['Area', 'Item', 'Year', 'Value']].copy()
balance2.Value  = balance2.Value *1000 #since the unit is 1000 tonnes

balance2 = balance2.groupby(['Area', 'Item', 'Year']).sum().reset_index()
balance2.rename(columns =  {'Value':'FoodInTonnes'}, inplace=True)
balance2 = balance2.dropna()
# balance2.head()

################Done food balance for humans

#search on Albania, for all the foods possible 

####################################### do trade -> import export 
# trade.Element.unique().tolist() 
# ==> ['Export Quantity', 'Export Value', 'Import Quantity', 'Import Value']
backupTrade = trade.copy()
trade.rename(columns = {'Reporter Countries': 'Area'}, inplace=True)
#kludge: we are filtering out the non tonness values 
trade = trade.loc[trade['Unit'] == 'tonnes'].copy()
#get import quantity, get export quantity, make 2 seperate columns
tradeImport = trade.loc[trade.Element == 'Import Quantity'].copy()
tradeExport = trade.loc[trade.Element == 'Export Quantity'].copy()
#remove unnessary columns 
tradeImport = tradeImport[['Year', 'Area', 'Item', 'Value' ]].copy()
tradeExport = tradeExport[['Year', 'Area', 'Item', 'Value' ]].copy()

#groupby to get total 
tradeImport = tradeImport.groupby(['Year', 'Area', 'Item']).sum().reset_index()
tradeImport.rename(columns = {'Value':'Import'}, inplace = True)
tradeExport = tradeExport.groupby(['Year', 'Area', 'Item']).sum().reset_index()
tradeExport.rename(columns = {'Value':'Export'}, inplace = True)
#join import-exports
tradeJoin = pd.merge(left=tradeImport, right = tradeExport, how = 'left', on =['Year', 'Area', 'Item'])
tradeJoin = tradeJoin.dropna()

################Done import export

####################################### do supply -> quantities
backupSupply = supply.copy()
#for learning about the datasets
# backupSupply.Element.unique().tolist()
# backupSupply.Unit.unique().tolist()
#filter out other units
befUnitsSupply = supply.Unit.unique().tolist()   #don't use this unless you have gotten "supply" df fresh
supply2 = supply.loc[supply['Element'] == 'Food supply quantity (tonnes)'].copy()
aftUnitsSupply = supply2.Unit.unique().tolist()
#######supply.Element.unique().tolist()  - only has food supply qty 
supply2 = supply2[['Year', 'Area', 'Item', 'Value' ]].copy()
supply2 = supply2.groupby(['Year', 'Area', 'Item']).sum().reset_index()
supply2 = supply2.dropna()
supply2.rename(columns = {'Value':'Qty'}, inplace=True)
# supply2.head()

################Done quantity

#should have year, area, totalValue grouped by the year, and the country, 
# trade2 = [['Reporter Countries']]

#kludge: make the population for the country remove it from top -> then attach it
#as a column joining on the year/year code values 
print('done')
```

```python
import tensorflow as tf
from tensorflow import keras
import sys
import pandas as pd
import pathlib as pl
import numpy as np

df =rObj('abc')
a =np.array(df.x.tolist())
b =np.array(df.y.tolist())

x = a
y = b
# x,y = normalizeInputs(a, b)

x_train = x[:9500].reshape(9500, 1)
y_train = y[:9500].reshape(9500, 1)

x_test = x[9500:].reshape(10000-9500, 1)
y_test = y[9500:].reshape(10000-9500, 1)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)


#plot data


model = keras.Sequential([
#     keras.layers.Dense(units=1,input_shape=(1,))
    keras.layers.Dense(1,input_shape=(1,)),
    keras.layers.Dense(128, activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.relu),#relu),
    keras.layers.Dense(1)
])


model.compile(optimizer=tf.train.AdamOptimizer(), #'sgd',#
              loss='mean_squared_error',
              metrics=['accuracy'])


model.fit(x_train, y_train, epochs=15)
test_loss, test_acc = model.evaluate(x_test, y_test)

# print('Test accuracy:', test_acc)
print('Test Loss', test_loss)

# https://stats.stackexchange.com/questions/284189/simple-linear-regression-in-keras

W, b = model.layers[0].get_weights()
print('Weights=', W, '\nbiases=', b)


x_old =np.array(df.x.tolist())
y_real =np.array(df.y.tolist())
y_new = model.predict(x_old)

new_df = pd.DataFrame({'x':x_old.flatten().tolist(), 'y':y_real.flatten().tolist(), 'y_pred':y_new.flatten().tolist()})
pObj(new_df, 'testing1')

y_new2 = model.predict(x_test)
new_d2 = pd.DataFrame({'x':x_test.flatten().tolist(), 'y':y_test.flatten().tolist(), 'y_pred':y_new2.flatten().tolist()})
pObj(new_d2, 'testing2')
```

```python
from sklearn.preprocessing import StandardScaler


def normalizeInputs(x, y):
    x =x.reshape(len(x),1)
    y =y.reshape(len(y),1)

    sc = StandardScaler()
    x = sc.fit_transform(x)
    y = sc.fit_transform(y)
    return x,y
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

import numpy as np
np.random.seed(1337)  # for reproducibility
 

# create some data
X = np.linspace(-1, 1, 200)
np.random.shuffle(X)    # randomize the data
Y = 0.5 * X + 2 + np.random.normal(0, 0.05, (200, ))



model = Sequential()
model.add(Dense(units=1, input_dim=1))
model.compile(loss='mse', optimizer='sgd')

# train test split
X_train, Y_train = X[:160], Y[:160]     
X_test, Y_test = X[160:], Y[160:]  




print('Training -----------')
for step in range(301):
    cost = model.train_on_batch(X_train, Y_train)
    if step % 100 == 0:
        print('train cost: ', cost)

print('\nTesting ------------')
cost = model.evaluate(X_test, Y_test, batch_size=40)
print('test cost:', cost)
W, b = model.layers[0].get_weights()
print('Weights=', W, '\nbiases=', b)
```
